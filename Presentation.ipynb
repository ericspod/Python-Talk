### Python 

* A general purpose dynamically-typed interpreted object-based multi-paradigm programming language
* Supports a broad range of programming methodologies from object-orientation to functional programming
* Designed to be an easy language to pick up, experiment with, and create simple scripts
* A "high level" language meaning a lot of complexity is handled for you by the interpreter (eg. memory management, compilation)

### Interpreted Code
* Python is an **interpreted language**, ie. code is read by an interpreter (or virtual machine, VM) which executes the code
* VM at runtime reads input code files, compiles to an intermediate form (bytecode), loads libraries, and then runs the compiled code through its machinery that emulates in a way a physical CPU
* **CPython** is the standard one distributed with Python
* Slower than compiled code but easier and quicker to go through the program->run->update development cycle
* Interpreter can be embedded in other apps, used remotely on servers, integrated with documents (eg. this one), etc

### Control Flow: Functions

      S
* A function begins with `def`, its name and arguments

### Control Flow: Exceptions
* The `raise` statement raises an exception and causes control flow to jump to the nearest appropriate `except` block
* "nearest" is defined as the first `except` block with the type of exception thrown that is associated with the `try` block most recent in the call stack
* Raising an exception means something has gone wrong and your code cannot function correctly, so someone else should clean up the mess
* This is a way of avoiding further errors, reporting where error occurred, and giving insight into how the error was arrived at

* `try` and `except` are code blocks what must appear together
* Optional code blocks `else` and `finally` can be used to trigger code when exceptions are not raised, or after they are handled

```python
try:
    raise RuntimeError, 'Something went wrong'
    print 'Unreachable'
except RuntimeError as e:
    print 'RuntimeError:',e
except Exception as e:
    print 'Exception: ',e
else:
    print 'No exception!'
finally:
    print 'Done'
```

* Without `try/except` blocks the exception propagates to the top of the call stack
* In script programs this causes the program to exit, for interactive shells you get a stack trace but then can continue
* An exception can be reraised in an `except` block with the statement `raise` without any values. 

```python
def divide(a,b):
    return a/b
print divide(3,0)
```

### Assertions
    assert C, S
* Statement checks `C`, if `False` an `AssertionError` is thrown with description `S`
* Meant for debugging and checking conditions that shouldn't ever occur, rather than validating input or other checks correct code must make
* Very useful as a sanity check in various places in code
* Also important for making explicit certain conditions in code that aren't obvious to the reader

* For example consider the greatest common divisor algorithm
* Takes 2 numbers a and b and computes the largest value both can be divided by evenly

```python
def gcd(a, b):
    # Do 'a' and 'b' have acceptable values?
    
    # Does this loop exit? How can we check?
    while b: 
        a, b = b, a % b # a % b is a modulo b, ie. the remainder of a//b
        
    # Is 'a' actually a divisor of the original values 'a' and 'b'?
    return a 

print gcd(16, 4), gcd(18, 4), gcd(7, 4)
```

```python
def gcd(a, b):
    if a < 0 or not isinstance(a, (int, long)):
        raise ValueError, "Argument 'a' must be positive integer"
    if b < 0 or not isinstance(b, (int, long)):
        raise ValueError, "Argument 'b' must be positive integer"

    a1, b1 = a, b # need to preserve original values for 'a' and 'b'
    while b1:
        assert (a1 % b1) < b1, "a1 % b1 should be < 'b1', loop might not exit"
        a1, b1 = b1, a1 % b1

    assert a % a1 == 0 and b % a1 == 0, "'a1' is not a divisor of 'a' and 'b'"
    return a1

print gcd(16,4), gcd(18,4), gcd(7,4)
print gcd(16,4.5)
```

### Objects
* Key feature to Python is that everything is an object (specifically first class objects)
* In this sense the architecture is object-based
* Objects in the most general sense are programming concepts that encapsulate state and operations, ie. data and routines that manipulate it
* Python is also object-oriented in that its object model supports subtyping, polymorphism, as well as encapsulation
* Object also have a concept of unique identity, type, and lifetime

* The form of objects (ie. type) is defined by a class, which provides a mechanism for creating instances of it (ie. objects), defining data and routines
* A simple type that represents a square area:

```python
class Area(object):
    def __init__(self, width, height): # constructor for object
        self.width = width # declare members by assignment
        self.height = height
    def area(self): # method of class
        return self.width * self.height
    
a=Area(10, 12) # a is an instance of Area
print id(a), type(a), a.width, a.height, a.area()
```

```python
b=Area(7,8) # different instance
print id(b), type(b), b.width, b.height, b.area()
```

```python
a=b # a and b now refer to the same object
print id(a), id(b)
```

* Classes are defined with the `class` statement, which gives the class name and any other inherited classes
* Constructors are always called `__init__` and are used to setup the object's state
* A routine of a class is called a method and can only access the members of the instance attached to it
* The `self` value always refers to the current object
* An instance of the class is created by calling the class name which will call `__init__` and is given a copy of the members the class defines
* Variables always store references to objects, that is they point to objects rather than store them

### Inheritance
* Inheritance is a key reuse and genericity feature of object-orientation:

```python
class Rect(Area):
    def __init__(self,x,y,width,height):
        Area.__init__(self,width,height)
        self.x=x
        self.y=y
        
    def inRect(self,x,y):
        return self.x <= x <= (self.x + self.width) and self.y <= y <= (self.y + self.height)
    
r=Rect(10,10,20,20)
print r.inRect(12,13),r.inRect(12,33), r.area()
print isinstance(r,Rect), isinstance(r,Area)
```

* Rect gets all of the members of Area plus defines its own (inheritance)
* Instances of Rect are also instances of Area (an aspect of polymorphism)
* The constructor from Area is called in that for Rect, this usually is necessary
* Rect could inherit from multiple classes, this creates problems if two classes have members with the same name

### Overridding
* An inherited method can be overridden (replaced) by the inheriting type:

```python
class Spam(object):
    def name(self):
        return 'Spam'
    def speak(self):
        print 'I am '+self.name()
        
class Eggs(Spam):
    def name(self):
        return 'Eggs'
    
s=Spam()
s.speak()
e=Eggs()
e.speak()
```

### Operators
* Methods can define behaviour when operators are used:

```python
class Area(object):
    def __init__(self, width, height): 
        self.width = width 
        self.height = height
    def __add__(self,a):
        return Area(self.width+a,self.height+a) # creates a new object
        
a=Area(10,10)
b=a+5
print a.width,b.width
```

* Other opeators for -, /, *, %, +=, -=, and, or, etc.
* Methods can accept multiple types and return different values
* Very important to use overloads for functionality that makes sense for the operator used, don't use `__add__` to implement area calculations

### Callable
* Everything's an object, even functions and methods
* An object is callable if it provides a `__call__` method which provides the code for when `()` brackets are used

```python
class Callable(object):
    def __call__(self,a):
        print a    
c=Callable()
c(123)
```

* Declaring functions like creating objects with a `__call__`:

```python
def CFunc(a):
    print a
CFunc(123)
print CFunc, CFunc.__call__
```
