### Control Flow: Exceptions
* The `raise` statement raises an exception and causes control flow to jump to the nearest appropriate `except` block
* "nearest" is defined as the first `except` block with the type of exception thrown that is associated with the `try` block most recent in the call stack
* Raising an exception means something has gone wrong and your code cannot function correctly, so someone else should clean up the mess
* This is a way of avoiding further errors, reporting where error occurred, and giving insight into how the error was arrived at

* `try` and `except` are code blocks what must appear together
* Optional code blocks `else` and `finally` can be used to trigger code when exceptions are not raised, or after they are handled

```python
try:
    raise RuntimeError, 'Something went wrong'
    print 'Unreachable'
except RuntimeError as e:
    print 'RuntimeError:',e
except Exception as e:
    print 'Exception: ',e
else:
    print 'No exception!'
finally:
    print 'Done'
```

* Without `try/except` blocks the exception propagates to the top of the call stack
* In script programs this causes the program to exit, for interactive shells you get a stack trace but then can continue
* An exception can be reraised in an `except` block with the statement `raise` without any values. 

```python
def divide(a,b):
    return a/b
print divide(3,0)
```

### Assertions
    assert C, S
* Statement checks `C`, if `False` an `AssertionError` is thrown with description `S`
* Meant for debugging and checking conditions that shouldn't ever occur, rather than validating input or other checks correct code must make
* Very useful as a sanity check in various places in code
* Also important for making explicit certain conditions in code that aren't obvious to the reader

* For example consider the greatest common divisor algorithm
* Takes 2 numbers a and b and computes the largest value both can be divided by evenly

```python
def gcd(a, b):
    # Do 'a' and 'b' have acceptable values?
    
    # Does this loop exit? How can we check?
    while b: 
        a, b = b, a % b # a % b is a modulo b, ie. the remainder of a//b
        
    # Is 'a' actually a divisor of the original values 'a' and 'b'?
    return a 

print gcd(16, 4), gcd(18, 4), gcd(7, 4)
```

```python
def gcd(a, b):
    if a < 0 or not isinstance(a, (int, long)):
        raise ValueError, "Argument 'a' must be positive integer"
    if b < 0 or not isinstance(b, (int, long)):
        raise ValueError, "Argument 'b' must be positive integer"

    a1, b1 = a, b # need to preserve original values for 'a' and 'b'
    while b1:
        assert (a1 % b1) < b1, "a1 % b1 should be < 'b1', loop might not exit"
        a1, b1 = b1, a1 % b1

    assert a % a1 == 0 and b % a1 == 0, "'a1' is not a divisor of 'a' and 'b'"
    return a1

print gcd(16,4), gcd(18,4), gcd(7,4)
print gcd(16,4.5)
```

### Objects
* Key feature to Python is that everything is an object (specifically first class objects)
* In this sense the architecture is object-based
* Objects in the most general sense are programming concepts that encapsulate state and operations, ie. data and routines that manipulate it
* Python is object-oriented in that its object model supports subtyping, polymorphism, as well as encapsulation
* Object also have a concept of unique identity, type, and lifetime

* The form of objects (ie. their type) is defined by a class, which provides a mechanism for creating instances of it, defining data, and defining routines
* A simple type that represents a square area:

```python
class Dimension(object):
    def __init__(self, width, height): # constructor for object
        self.width = width # declare members by assignment
        self.height = height
    def area(self): # method of class
        return self.width * self.height
    
a=Dimension(10, 12) # a is an instance of Dimension
print id(a), type(a), a.width, a.height, a.area()
```

```python
b=Dimension(7,8) # different instance
print id(b), type(b), b.width, b.height, b.area()
```

```python
a=b # a and b now refer to the same object
print id(a), id(b)
```

* Classes are defined with the `class` statement, which gives the class name and any other inherited classes
* Constructors are always called `__init__` and are used to setup the object's state
* Calling the class name creates an instance with a copy of the members the class defines, then `__init__` is called
* Methods are class routines, they can only access the members of the associated instance, its arguments, and global variables
* The `self` value always refers to the current object
* Variables always store references to objects, that is they point to objects rather than store them

### Inheritance
* Inheritance is a key reuse and genericity feature of object-orientation:

```python
class Rect(Dimension):
    def __init__(self,x,y,width,height):
        Dimension.__init__(self,width,height)
        self.x=x
        self.y=y
        
    def inRect(self,x,y):
        return self.x <= x <= (self.x + self.width) and \
            self.y <= y <= (self.y + self.height)
    
r=Rect(10,10,20,20)
print r.inRect(12,13),r.inRect(12,33), r.area()
print isinstance(r,Rect), isinstance(r,Dimension)
```

* Rect gets all of the members of Dimension plus those it defines (inheritance)
* Instances of Rect are also instances of Dimension (an aspect of polymorphism)
* The constructor from Dimension is called in that for Rect, this usually is necessary
* Rect could inherit from multiple classes, this creates problems if two classes have members with the same name

### Overridding
* An inherited method can be overridden (replaced) by the inheriting type:

```python
class Spam(object):
    def name(self):
        return 'Spam'
    def speak(self):
        print 'I am '+self.name()
        
class Eggs(Spam):
    def name(self):
        return 'Eggs'
    
s=Spam()
s.speak()
e=Eggs()
e.speak()
```

### Iterator Design Pattern
* An iterable is an object which produces a successive value on request from either a collection or by generation
* Abstracts away how one traverses a data structure, creates a sequence of values from a specification, etc

```python
r=range(5)
print r
it=iter(r)
print it
print next(it)
print next(it)
```

* `iter` calls the method `__iter__` of its argument
* Object with `__iter__` returning an iterator is termed Iterable
* Iterators are objects with a `next` method
* `next` is called by `next()` and is responsible for returning the next value in sequence, or raising `StopIteration` when done

```python
class mylistiterator(object):
    def __init__(self,lst):
        self.pos=0
        self.lst=lst
    def next(self): # returns successive values
        if self.pos<len(self.lst):
            self.pos+=1
            return self.lst[self.pos-1]
        raise StopIteration
            
it=mylistiterator(range(5))
print next(it), next(it), next(it), next(it), next(it)
print next(it)
```

* A simple data structure can be defined as iterable with an `__iter__` which returns an iterator appropriate for its contents:

```python
class mylist(object):
    def __init__(self,lst):
        self.lst=lst
    def __iter__(self):
        return mylistiterator(self.lst)
    
m=mylist(['spam','eggs',1,2])
print tuple(m) 
```

* `tuple` takes each item from its argument and puts it into a tuple

* We saw the `for` statement earlier, this uses iterables and simply assigns each successive value to the loop variable

```python
for i in range(5):
    print i
    
# roughly equivalent to the above
it=iter(range(5))
while True:
    try:
        i=next(it)
    except StopIteration:
        break
        
    print i
```

* Values can be defined algorithmically:

```python
class myrange(object):
    def __init__(self,start,stop,inc):
        self.val=start
        self.stop=stop
        self.inc=inc
    def __iter__(self): # both iterator and iterable
        return self
    def next(self):
        if self.val>=self.stop:
            raise StopIteration
        v=self.val
        self.val+=self.inc
        return v

r=myrange(0,10,2)
print tuple(r)
```

### Generators
* An easier way of doing this is with a generator routine
* Generators use the `yield` statement to break out of control flow to produce a value then continuing execution
* Return value is an iterable
* An instance of a coroutine

```python
def myrange(start,stop,inc):
    while start<stop:
        yield start # stop here and return control flow until next() is used
        start+=inc
        
r=myrange(0,10,2)
print r
print next(r),next(r)
print tuple(r)
```

### Matplotlib
 * Numpy provides the mathematical infrastructure, Matplotlib provides the graphing and visualization
 * Again deliberately Matlab-like
 * Plots are made by creating objects and configuring them through methods
 * Alternatively __`pyplot`__ module provides a state machine interface like Matlab, easier to use for simple tasks

```python
import matplotlib.pyplot as plt
import numpy as np
# puts the plot here in the notebook
%matplotlib inline

x=np.arange(0,np.pi*4,np.pi/20)
plt.plot(x,np.sin(x),label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
plt.title('Example Plot')
plt.legend()
plt.show()
```

 * Alternative object way of doing the same thing:

```python
fig = plt.figure(1) # create a Figure object 
ax = fig.add_subplot(111) # create an Axes object
print fig,ax

ax.plot(x,np.sin(x),label='sin(x)')
ax.plot(x,np.cos(x),label='cos(x)')
ax.set_title('Example Plot')
ax.legend()
plt.show()
```

 * A plot can have multiple subplots:

```python
plt.figure(figsize=(20,5)) # 12x6 inch plots

plt.subplot(121) # 1 row, 2 colums, plot 1
plt.plot(x,np.sin(x),label='sin(x)')
plt.title('Sin')

plt.subplot(122)  # 1 row, 2 colums, plot 2
plt.plot(x,np.cos(x),label='cos(x)')
plt.title('Cos')
```

 * Settings for scale, ticks, grid, labels, annotations, etc.

```python
y=np.sort(np.random.normal(0.5, 0.5, 100))
plt.figure(figsize=(20,5))
plt.ylabel('$\log_{10}$ Scale',fontsize=20,color='red') # latex in labels
plt.yscale('log') # logarithmic scale
plt.xticks(np.arange(0,len(y),5)) # ticks every 5
plt.grid(True)
plt.xlim(10,90) # limit x axis to range [10,90]
plt.annotate("What's that?!",xy=(25,0.125),xytext=(37,0.1),\
             fontsize=20,arrowprops={'shrink':0.05})
plt.plot(np.arange(len(y)),y,'m--',linewidth=3) # magenta dashed line
```
