```python
%matplotlib inline  
import numpy as np
import matplotlib.pyplot as plt
def middist(*ndims):
    mids=[(n-1)/2.0 for n in ndims]
    f=lambda *inds: sum((i-m)**2 for i,m in zip(inds,mids))**0.5
    return np.fromfunction(f,ndims)
f = plt.figure(figsize = (7,7))
ax = f.add_subplot(111)
ax.imshow(middist(50,50), interpolation='none')
```

```python
mag=lambda *vals: sum(v*v for v in vals)**0.5

f = plt.figure(figsize = (7,7))
ax = f.add_subplot(111)
ax.imshow(np.fromfunction(mag,(50,50)), interpolation='bilinear')
```

```python
func=lambda x,y: np.cos(x/25.0*np.pi)*np.sin(y/25.0*np.pi)

f = plt.figure(figsize = (7,7))
ax = f.add_subplot(111)
ax.imshow(np.fromfunction(func,(50,50)), interpolation='catrom')
```
